```python
import pandas as pd
import os
```

```python
customers = pd.read_csv('../staging_1/customers/customers.csv')
customers.head()
```

```python
stores = pd.read_csv('../staging_1/stores/stores.csv')
stores
```

```python
customers['local'] = customers['city'].isin(stores['city'].unique())
customers.head()
```

```python
#Reorder columns to be more readable:
desired_column_order = ['customer_id', 'first_name', 'last_name', 'email', 'street', 'city', 'state', 'local', 'zip_code', 'Extraction_Date_Time', 'Source']
customers = customers.reindex(columns=desired_column_order)
customers.head()
```

```python
file_path = '../Staging_2/customers/customers.csv'

os.makedirs(os.path.dirname(file_path), exist_ok=True)
customers.to_csv(file_path, index = False)
```
