```python
import pandas as pd
import os
```

```python
order_items = pd.read_csv('../staging_1/order_items/order_items.csv')
order_items
```

```python
exchange = pd.read_csv('../staging_1/exchange/exchange.csv')
exchange
```

```python
order_items = pd.merge(order_items,exchange,how='cross')
order_items.head()
```

```python
order_items = order_items.drop(['Extraction_Date_Time_y', 'Source_y', 'timestamp', 'base'], axis=1)
order_items.head()
```

```python
order_items.rename(columns={'Extraction_Date_Time_x': 'Extraction_Date_Time', 'Source_x' : 'Source'}, inplace=True)
order_items
```

```python
order_items['list_price_egp'] = order_items['list_price'] * order_items['EGP']
order_items
```

```python
#Reorder columns to be more readable:
desired_column_order = ['order_id', 'item_id', 'product_id', 'quantity', 'list_price', 'EGP', 'list_price_egp','discount', 'Extraction_Date_Time', 'Source']
order_items = order_items.reindex(columns=desired_column_order)
order_items
```

```python
file_path = '../Staging_2/order_items/order_items.csv'

os.makedirs(os.path.dirname(file_path), exist_ok=True)
order_items.to_csv(file_path, index = False) 
```
