```python
import pandas as pd
import os
```

```python
orders = pd.read_csv('../staging_2/orders/orders.csv')
orders.head()
```

```python
products = pd.read_csv('../staging_2/products/products.csv')
products.head()
```

```python
order_items = pd.read_csv('../staging_2/order_items/order_items.csv')
order_items.head()
```

```python
orders_items = pd.merge(orders, order_items , on='order_id' , how='inner')
orders_items.head()
```

```python
orders_products_details = pd.merge(orders_items , products , on='product_id' , how='inner')
orders_products_details.head()
```

```python
columns_to_drop = ['Extraction_Date_Time_x', 'Source_x', 'Extraction_Date_Time_y', 'list_price_y', 'Source_y', 'Extraction_Date_Time', 'Source']
# Drop the columns
orders_products_details.drop(columns=columns_to_drop, inplace=True)
```

```python
orders_products_details.rename(columns={'list_price_x': 'list_price'}, inplace=True)
```

```python
orders_products_details.head()
```

```python
order_status_lookup =  pd.read_csv('../Staging_2/order_status_lookup/order_status_lookup.csv')
order_status_lookup.head()
```

```python
orders_products_details = pd.merge(orders_products_details, order_status_lookup, left_on='order_status', right_on='order_status_id', how='inner')
orders_products_details.head()
```

```python
orders_products_details.info()
```

```python
columns_to_drop = ['order_status_description_y', 'order_status_id_y', 'order_status_id_x', 'order_status']
# Drop the columns
orders_products_details.drop(columns=columns_to_drop, inplace=True)
```

```python
orders_products_details.rename(columns={'order_status_description_x': 'order_status'}, inplace=True)
```

```python
#Reorder columns to be more readable:
desired_column_order = ['order_id', 'customer_id', 'store_id',
       'staff_id', 'item_id', 'product_id', 'product_name', 'brand_id', 'category_id',
       'model_year','quantity', 'list_price', 'EGP', 'list_price_egp', 'discount' ,'order_status', 
       'order_date', 'required_date','shipped_date', 'late_delivery', 'latency_days', 'weekend']

orders_products_details = orders_products_details.reindex(columns=desired_column_order)
orders_products_details.info()
```

```python
file_path = '../Information Mart/orders_products_details/orders_products_details.csv'

os.makedirs(os.path.dirname(file_path), exist_ok=True)
orders_products_details.to_csv(file_path, index = False) 
```
