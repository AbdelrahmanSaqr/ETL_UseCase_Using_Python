```python
import pandas as pd
import sqlalchemy as db
import os
```

```python
con = db.create_engine('postgresql://postgres:root@127.0.0.1/bikestore')
```

```python
query = """
SELECT *
FROM general.orders
"""
orders = pd.read_sql(query, con)
orders.head()
```

```python
orders['Extraction_Date_Time'] = pd.Timestamp.now()
orders["Source"] = "DB"

file_path = './Landing/orders/orders.csv'
os.makedirs(os.path.dirname(file_path), exist_ok=True)
orders.to_csv(file_path, index = False) 
```
